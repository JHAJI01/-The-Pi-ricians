```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/JHAJI01/-The-Pi-ricians/main/covid_data.csv"
df = pd.read_csv(url)
```

```python
df['Total Cases Clean'] = df['Total Cases'].str.replace(',', '').astype(float)
```

```python
df_clean = df.dropna(subset=['Total Cases Clean'])
```

```python
df_clean['FirstDigit'] = df_clean['Total Cases Clean'].astype(str).str[0].astype(int)
```

```python
actual = df_clean['FirstDigit'].value_counts(normalize=True).sort_index()
```

```python
benford = [np.log10(1 + 1/d) for d in range(1, 10)]
```

```python
digits = np.arange(1, 10)
expected = benford
actual = actual.reindex(digits, fill_value=0)
```

```python
plt.figure(figsize=(12, 7))

width = 0.35
x = np.arange(len(digits))

plt.bar(x - width/2, expected, width, label='Expected (Benford)', color='skyblue', edgecolor='black')
plt.bar(x + width/2, actual, width, label='Actual (COVID Total Cases)', color='orange', edgecolor='black')

plt.xlabel('First Digit', fontsize=14, weight='bold')
plt.ylabel('Frequency', fontsize=14, weight='bold')
plt.title("Benford's Law Analysis – COVID-19 Total Cases", fontsize=16, weight='bold')

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.xticks(x, digits, fontsize=12)
plt.yticks(fontsize=12)

plt.legend(fontsize=12)

for i, (e, a) in enumerate(zip(expected, actual)):
    plt.text(x[i] - width/2, e + 0.01, f'{e:.2f}', ha='center', va='bottom', fontsize=10, color='blue', weight='bold')
    plt.text(x[i] + width/2, a + 0.01, f'{a:.2f}', ha='center', va='bottom', fontsize=10, color='red', weight='bold')

plt.tight_layout()

plt.show()
```
